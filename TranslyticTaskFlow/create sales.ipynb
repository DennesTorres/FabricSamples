```python
from pyspark.sql import SparkSession
from pyspark.sql.types import StructType, StructField, DateType, FloatType
from datetime import date

# Create dummy data for May 16–20, 2025
# Outlier: May 17 is a dip, May 19 is a spike
sales_data = [
    (date(2025, 5, 16), 1500.0),  # Normal day
    (date(2025, 5, 17), 500.0),   # Outlier low
    (date(2025, 5, 18), 1600.0),  # Normal day
    (date(2025, 5, 19), 3000.0),  # Outlier high
    (date(2025, 5, 20), 1400.0),  # Normal day
]

# Define schema
schema = StructType([
    StructField("sale_date", DateType(), False),
    StructField("total_sales", FloatType(), False)
])

# Create DataFrame
df = spark.createDataFrame(sales_data, schema=schema)

# Optional: Show the data
df.show()

# Save the table to the Lakehouse (replace if it already exists)
df.write.mode("overwrite").saveAsTable("salestotals")

print("Table 'salestotals' written to the Lakehouse.")
```
