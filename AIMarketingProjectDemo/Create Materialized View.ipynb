```python
%%sql
Create MATERIALIZED LAKE VIEW Silver.AmazonSentimentNew AS
SELECT
    CAST(SUBSTRING(r.__filepath__, -8, 4) AS INT) AS Id,
    a.reviewerName,
    a.overall,
    a.reviewText,
    a.day_diff,
    a.helpful_yes,
    a.helpful_no,
    a.total_vote,
    a.score_pos_neg_diff,
    a.score_average_rating,
    a.wilson_lower_bound,
    (r.PositiveConfidenceScore - r.NegativeConfidenceScore) AS sentimentScore,
    r.Sentiment
FROM amazon a
INNER JOIN AmazonReviews r
    ON a.`key` = CAST(SUBSTRING(r.__filepath__, -8, 4) AS INT)
WHERE r.Classification = 'Document level'
```
