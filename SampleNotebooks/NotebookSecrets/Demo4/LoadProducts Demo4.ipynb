```python
categories=(1,2,3,4)
```

```python
DAG = {
    "activities": [
        {
            "name": f"CategoryId{cat}", 
            "path": "Load Single Category Demo 4", 
            "timeoutPerCellInSeconds": 300,
            "args": {"function_url": f"{function_url}?CategoryId={cat}",
                     "cat": cat,
                    "executionId": executionId},
        }
        for cat in categories
                ],
    }

result=mssparkutils.notebook.runMultiple(DAG)
```

```python

import json

results=[]

for cat in categories:
    callname=f"CategoryId{cat}"
    exitVal=result[callname]['exitVal']

    if exitVal==None:
        continue

    data=json.loads(exitVal)
    df=spark.createDataFrame(data)
    results.append(df)
```

```python
mainDF=common.joinResults(results)

display(mainDF)
```
